```python
file_path = 'Cleaned_data.pkl'
df = pd.read_pickle(file_path)#.sample(n=10_000_000, random_state=42)
```

```python
# Features considered
features = [
'Wind.speed.me',
'Wind.speed.sd',
'Front.bearing.temp.me',
'Rear.bearing.temp.me',
'Stator1.temp.me',
'Nacelle.ambient.temp.me',
'Nacelle.temp.me',
'Transformer.temp.me',
'Gear.oil.temp.me',
'Gear.oil.temp.me.1',
'Top.box.temp.me',
'Hub.temp.me',
'Conv.Amb.temp.me',
'Rotor.bearing.temp.me',
'Transformer.cell.temp.me',
'Motor.axis1.temp.me',
'Motor.axis2.temp.me',
'Motor.axis3.temp.me',
'CPU.temp.me',
'Rotor.speed.me',
'Blade.ang.pitch.pos.A.me',
'Blade.ang.pitch.pos.B.me',
'Blade.ang.pitch.pos.C.me',
'Gear.oil.inlet.press.me',
'Gear.oil.pump.press.me',
'Drive.train.acceleration.me',
'Tower.Acceleration.x',
'Tower.Acceleration.y'
]


output_feature = 'Power.me'

# Only consider the first turbine
df = df[df['turbine'] == 2]

df = df.dropna(subset=features + [output_feature])
df = df[features + [output_feature]]
# df = df.sample(n=100000, random_state=42) 
```

```python
def preprocess_pipeline_power(df):
 scaler = StandardScaler()
 power_transform = PowerTransformer(method='yeo-johnson')
 pipeline_power = Pipeline([('s', scaler), ('p', power_transform)])
 res_power = pipeline_power.fit_transform(df)
 x_train, x_test, y_train, y_test = train_test_split(res_power[:,0:-1], res_power[:,-1], test_size = 0.2, random_state=123)
 return(x_train, x_test, y_train, y_test)
```

```python
from sklearn.model_selection import train_test_split

def load_dataset_train_validation_test_split(df, features, output_feature):
    # Dataset to train the autoencoder (full dataset)
    X = df[features]
    y = df[output_feature]

    # Split the data into training, validation, and test sets
    # First, split into training (70%) and temp (30%)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=12)

    # Standardize both X and y
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    return X_train_scaled, X_test_scaled, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = load_dataset_train_validation_test_split(df, features, output_feature)
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

Use completely separate neural networks (not sharing hidden layers) to model mean and std separately

```python
def NLL(y, distr):
  return -distr.log_prob(y)

def norm(params):
  return tfd.Normal(loc=params[:,0:1], scale=1e-3 + tf.math.softplus(0.05 * params[:,1:2])) # both parameters are learnable

def lapl(params):
    loc = params[:, 0:1]
    scale = 1e-3 + tf.math.softplus(0.05 * params[:, 1:2])
    return tfd.Laplace(loc=loc, scale=scale)

def get_model(X):
    if(isinstance(X, pd.DataFrame)):
        Xlen = len(X.columns)
    else:
        Xlen = np.shape(X)[1]
    input1 = Input(shape=(X.shape[1], X.shape[2])) # 13 for boston housing and 8 for california housing data
    conv1 = Conv1D(filters=32, kernel_size=3, activation='relu')(input1)
    conv2 = Conv1D(filters=32, kernel_size=3, activation='relu')(input1)
#     pool1 = MaxPooling1D(pool_size=2)(conv1)
#     pool2 = MaxPooling1D(pool_size=2)(conv2)
    flat1 = Flatten()(conv1)
    flat2 = Flatten()(conv2)

    hidden1 = Dense(32, activation='relu', name='dense_1')(flat1) # 32 or 8
    hidden2 = Dense(32, activation='relu', name='dense_2')(flat2)
    out1 = Dense(1, activation='relu', name='out_1')(hidden1) # out1 is mean
    out2 = Dense(1, activation='relu', name='out_2')(hidden2) # out2 is std
    params = Concatenate()([out1,out2]) #C
    dist = tfp.layers.DistributionLambda(norm)(params)
    model = Model(inputs=input1, outputs=dist)
    model.compile(tf.keras.optimizers.Adam(learning_rate=0.001), loss=NLL)
    model_mean = Model(inputs=input1, outputs=out1)  # Access the mean directly
    model_std = Model(inputs=input1, outputs=out2)  # Access the std deviation directly
    model.summary()
    return(model, model_mean, model_std)

def evaluate_model(model, model_mean, model_std, X_data_test, y_data_test):
    y_out_mean = model_mean.predict(X_data_test)
    y_out_std = model_std.predict(X_data_test)
    y_out_mean_vals = y_out_mean.squeeze(axis=1)
    if(isinstance(y_data_test, pd.DataFrame)):
        y_test_true_vals = y_data_test.values.squeeze(axis=1)
    else:
        y_test_true_vals = y_data_test
    y_out_std_vals = y_out_std.squeeze(axis=1)
    neg_log_prob_array = []
    for elem in zip(y_out_mean_vals, y_test_true_vals, y_out_std_vals):
        predicted = elem[0]
        predicted_var = elem[2]
        true_val = elem[1]
        neg_log_prob = -1.0 * tfd.Normal(predicted, predicted_var).log_prob(true_val).numpy()
        neg_log_prob_array.append(neg_log_prob)
    return(neg_log_prob_array)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Define early stopping criteria
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Create and compile your model
model, model_mean, model_std = get_model(X_train)

# Train your model with early stopping
history = model.fit(X_train, y_train, batch_size=128, epochs=50, validation_split=0.1, callbacks=[early_stopping])
```

```python
# plot train and validation loss
plt.plot(history.history['loss'][1:])
plt.plot(history.history['val_loss'][1:])
plt.title('model train vs validation loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Calculate predictions and uncertainties for both training and test datasets
train_means = model_mean(X_train)  # This gives you the whole distribution
train_stddevs = model_std(X_train)
test_means = model_mean(X_test)
test_stddevs = model_std(X_test)

# Calculate the prediction intervals (e.g., 95% prediction intervals)
confidence_level = 0.95
z_score = 1.96  # For 95% prediction intervals (you can change this)

# Calculate the lower and upper bounds of the prediction intervals
test_lower = test_means - z_score * test_stddevs
test_upper = test_means + z_score * test_stddevs

# Calculate MSE, RMSE, and MAE for training dataset
train_mse = mean_squared_error(y_train, train_means)
train_rmse = np.sqrt(train_mse)
train_mae = mean_absolute_error(y_train, train_means)

# Calculate MSE, RMSE, and MAE for test dataset
test_mse = mean_squared_error(y_test, test_means)
test_rmse = np.sqrt(test_mse)
test_mae = mean_absolute_error(y_test, test_means)

# Print the results
print("Training MSE:", train_mse)
print("Training RMSE:", train_rmse)
print("Training MAE:", train_mae)

print("Test MSE:", test_mse)
print("Test RMSE:", test_rmse)
print("Test MAE:", test_mae)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_scatter_for_feature(X, y, model, feature_index, feature_name):
    # Number of samples to plot
    num_samples_to_plot = 1000

    # Randomly sample 1000 data points
    sample_indices = np.random.choice(X.shape[0], num_samples_to_plot, replace=False)
    X_samples = X[sample_indices, [feature_index]]
    y_samples = np.array(y)[sample_indices]

    # Calculate the predicted means for the samples
    y_model = model(X)
    y_means = np.array(y_model.mean())[sample_indices]

    # Create the scatter plot
    plt.figure(figsize=(12, 8))
    plt.scatter(X_samples, y_samples, label='True Values', alpha=0.4)
    plt.scatter(X_samples, y_means, label='Predicted Means', color='red', alpha=0.4)
    plt.xlabel(feature_name)
    plt.ylabel('Output')
    plt.legend()
    plt.title(f'Scatter Plot of True Values and Predicted Means ({feature_name})')
    plt.show()

plot_scatter_for_feature(X_test, y_test, model, feature_index=0, feature_name='Wind.speed.me')
plot_scatter_for_feature(X_test, y_test, model, feature_index=18, feature_name='Rotor.speed.me')
plot_scatter_for_feature(X_test, y_test, model, feature_index=1, feature_name='Front.bearing.temp.me')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 3: Bucketize Predictive Values
num_buckets = 10
hist, bin_edges = np.histogram(predictions, bins=num_buckets)

# Step 4: Plot Histograms for Each Bucket on Separate Subplots
fig, axes = plt.subplots(num_buckets, 1, figsize=(8, 12))

for i in range(num_buckets):
    bucket_mask = np.logical_and(predictions >= bin_edges[i], predictions <= bin_edges[i+1])
    bucket_residuals = residuals[bucket_mask.ravel()]
    
    axes[i].hist(bucket_residuals, bins=100, alpha=0.7)
    axes[i].set_title(f'Bucket {i+1}')
    axes[i].set_xlabel('Residuals')
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.subplots_adjust(top=0.9)
fig.suptitle('Residuals Histogram for Different Buckets of Predictive Values', fontsize=16)
plt.show()
```
