```python
def plot_means_variances(y_true, y_means, y_stddevs):
    min_vals = np.min([np.min(y_true), np.min(y_means)])
    max_vals = np.max([np.max(y_true), np.max(y_means)])

    plt.figure(figsize=(16, 6))

    # Plot means
    plt.subplot(1, 2, 1)
    plt.scatter(y_true, y_means, alpha=0.5, s=5)
    plt.plot([min_vals, max_vals], [min_vals, max_vals], 'k--')  # Add diagonal line
    plt.title('Means')
    plt.xlabel('True Values')
    plt.ylabel('Predicted Values')

    # Plot standard deviations
    plt.subplot(1, 2, 2)
    plt.scatter(y_true, y_stddevs, alpha=0.5, s=5)
    plt.title('Standard Deviation')
    plt.xlabel('True Values')
    plt.ylabel('Predicted Standard Deviation')

    plt.tight_layout()
    plt.show()

    
def evaluate_and_print_metrics(y_train, y_test, y_train_pred, y_test_pred, y_train_stddevs, y_test_stddevs, ci):
    z_value = stats.norm.ppf((1 + ci) / 2)
    
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

    train_mae = mean_absolute_error(y_train, y_train_pred)
    test_mae = mean_absolute_error(y_test, y_test_pred)

    train_lower_bound = y_train_pred - z_value * y_train_stddevs
    train_upper_bound = y_train_pred + z_value * y_train_stddevs

    test_lower_bound = y_test_pred - z_value * y_test_stddevs
    test_upper_bound = y_test_pred + z_value * y_test_stddevs

    train_within_interval = np.sum(np.logical_and(y_train.ravel() >= train_lower_bound, y_train.ravel() <= train_upper_bound))
    test_within_interval = np.sum(np.logical_and(y_test.ravel() >= test_lower_bound, y_test.ravel() <= test_upper_bound))

    train_percentage_within_interval = (train_within_interval / len(y_train.ravel())) * 100
    test_percentage_within_interval = (test_within_interval / len(y_test.ravel())) * 100

    # Calculate normalized mean squared error (NMSE)
    train_nrmse = 100 * np.sqrt(mean_squared_error(y_train, y_train_pred)) / (np.var(y_train)*len(y_train))
    test_nrmse = 100 * np.sqrt(mean_squared_error(y_test, y_test_pred)) / (np.var(y_test)*len(y_test))

    # Print the results
    print(f"Train RMSE: {train_rmse:.3f}")
    print(f"Test RMSE: {test_rmse:.3f}")
    print(f"Train MAE: {train_mae:.3f}")
    print(f"Test MAE: {test_mae:.3f}")
    print(f"Percentage of Train Data Points within {ci:.2f}% CI: {train_percentage_within_interval:.2f}%")
    print(f"Percentage of Test Data Points within {ci:.2f}% CI: {test_percentage_within_interval:.2f}%")
    print(f"Train Normalised MSE: {train_nrmse:.8f} %")
    print(f"Test Normalised MSE: {test_nrmse:.8f} %")
```

```python
def plot_residuals_with_confidence_intervals(timestamps, residuals, upper_limit, n_timestamps=500):
    # Plot residuals for the last n_timestamps data points
    plt.figure(figsize=(16, 6))

    # Calculate absolute residuals
    abs_residuals = np.abs(residuals[-n_timestamps:])

    # Plot absolute residuals
    plt.plot(timestamps.iloc[-n_timestamps:], abs_residuals, label='Test Residuals (Absolute Value)', color='orange')

    # Plot positive 95% confidence interval limit
    plt.fill_between(timestamps.iloc[-n_timestamps:], 0, upper_limit[-n_timestamps:], color='orange', alpha=0.2)

    # Mark points where residuals surpass the upper limit with red dotted lines
    surpass_upper_limit = abs_residuals > (upper_limit[-n_timestamps:])
    surpass_upper_limit_timestamps = np.array(timestamps.iloc[-n_timestamps:])[surpass_upper_limit]
    surpass_upper_limit_residuals = abs_residuals[surpass_upper_limit]

    plt.plot(surpass_upper_limit_timestamps, surpass_upper_limit_residuals, 'ro', linestyle='dotted', label='Surpasses Upper Limit')

    plt.axhline(0, color='black', linestyle='--', linewidth=1)  # Add a horizontal line at y=0 for reference

    plt.title('Absolute Residuals and Positive 95% Confidence Intervals')
    plt.xlabel('Timestamp')
    plt.ylabel('Absolute Residuals')
    plt.legend()
    plt.show()
```

```python
def plot_smoothed_residuals_with_smoothed_confidence_intervals(timestamps, residuals, upper_limit, window_size=5, n_timestamps=500):
    # Plot residuals for the last n_timestamps data points
    plt.figure(figsize=(16, 6))

    # Calculate absolute residuals
    abs_residuals = np.abs(residuals[-n_timestamps:])

    # Apply a moving average window to the absolute residuals
    smoothed_abs_residuals = pd.Series(abs_residuals).rolling(window=window_size, min_periods=1).mean()

    # Plot smoothed absolute residuals
    plt.plot(timestamps.iloc[-n_timestamps:], smoothed_abs_residuals, label=f'Test Residuals (Moving Average, Window={window_size})', color='orange')

    # Calculate and apply a moving average window to the positive 95% confidence interval limit
    smoothed_upper_limit = pd.Series(upper_limit[-n_timestamps:]).rolling(window=window_size, min_periods=1).mean()

    # Plot smoothed positive 95% confidence interval limit
    plt.fill_between(timestamps.iloc[-n_timestamps:], 0, smoothed_upper_limit, color='orange', alpha=0.2)

    # Reset indices for the comparison
    smoothed_abs_residuals.reset_index(drop=True, inplace=True)
    smoothed_upper_limit.reset_index(drop=True, inplace=True)

    # Mark points where smoothed residuals surpass the smoothed upper limit with red dotted lines
    surpass_upper_limit = smoothed_abs_residuals > smoothed_upper_limit
    surpass_upper_limit_timestamps = np.array(timestamps.iloc[-n_timestamps:])[surpass_upper_limit]
    surpass_upper_limit_residuals = smoothed_abs_residuals[surpass_upper_limit]

    plt.plot(surpass_upper_limit_timestamps, surpass_upper_limit_residuals, 'ro', linestyle='dotted', label='Surpasses Upper Limit')

    plt.axhline(0, color='black', linestyle='--', linewidth=1)  # Add a horizontal line at y=0 for reference

    plt.title(f'Smoothed Absolute Residuals and Smoothed Positive 95% Confidence Intervals (Window={window_size})')
    plt.xlabel('Timestamp')
    plt.ylabel('Smoothed Absolute Residuals')
    plt.legend()
    plt.show()
```

```python
def plot_binned_averages(y_pred, y_stddev, num_bins=50):
    # Create an array of bin edges
    bin_edges = np.linspace(min(y_pred), max(y_pred), num_bins + 1)

    # Calculate the means and standard deviations within each bin
    bin_means = []
    bin_stddevs = []

    for i in range(num_bins):
        mask = (y_pred >= bin_edges[i]) & (y_pred < bin_edges[i + 1])
        if np.any(mask):
            bin_means.append(np.mean(y_pred[mask]))
            bin_stddevs.append(np.mean(y_stddev[mask]))
        else:
            bin_means.append(0)  # Placeholder for empty bins
            bin_stddevs.append(0)

    # Create a scatter plot of binned averages
    plt.figure(figsize=(8, 6))
    plt.scatter(bin_means, bin_stddevs, c='b', marker='o', alpha=0.5)
    plt.title('Binned Averages of Means vs. Standard Deviations')
    plt.xlabel('Binned Mean')
    plt.ylabel('Average Standard Deviation')
    plt.grid(True)
    plt.show()
```

```python
def plot_scatter_for_feature(X, y, model, feature_index, feature_name):
    # Number of samples to plot
    num_samples_to_plot = 1000

    # Randomly sample 1000 data points
    sample_indices = np.random.choice(X.shape[0], num_samples_to_plot, replace=False)
    X_samples = X[sample_indices, [feature_index]]
    y_samples = np.array(y)[sample_indices]

    # Calculate the predicted means for the samples
    y_model = model(X)
    y_means = np.array(y_model.mean())[sample_indices]

    # Create the scatter plot
    plt.figure(figsize=(12, 8))
    plt.scatter(X_samples, y_samples, label='True Values', alpha=0.4)
    plt.scatter(X_samples, y_means, label='Predicted Means', color='red', alpha=0.4)
    plt.xlabel(feature_name)
    plt.ylabel('Output')
    plt.legend()
    plt.title(f'Scatter Plot of True Values and Predicted Means ({feature_name})')
    plt.show()
```

```python
def plot_confidence_interval_histogram(y_test_pred, y_test_std, y_test, bins=20):
    # Compute the lower and upper bounds of the confidence intervals based on Z-scores
    z_scores = np.array([norm.ppf(i/bins + (1-i/bins)/2) for i in range(1, bins+1)])

    # Initialize arrays to store percentages
    percentages_within_interval = []

    for z_score in z_scores:
        lower_bounds = y_test_pred.ravel() - z_score * y_test_std
        upper_bounds = y_test_pred.ravel() + z_score * y_test_std

        # Count how many data points are within the confidence interval
        is_within_interval = np.logical_and(y_test >= lower_bounds, y_test <= upper_bounds)
        num_within_interval = np.sum(is_within_interval)

        # Calculate the percentage of data points within the confidence interval
        percentage_within_interval = (num_within_interval / len(y_test)) * 100
        percentages_within_interval.append(percentage_within_interval)

    # Plot the histogram with percentages over confidence intervals
    plt.figure(figsize=(16, 6))
    bars = plt.bar(np.arange(1, bins+1), percentages_within_interval, color='skyblue', edgecolor='black')
    plt.xlabel('Confidence Intervals')
    plt.ylabel('Percentage within Interval')
    plt.title('Percentage of Data Points within Confidence Intervals')

    # Add numbers on top of each bar
    for bar, percentage in zip(bars, percentages_within_interval):
        plt.text(bar.get_x() + bar.get_width() / 2 - 0.15, bar.get_height() + 1, f'{percentage:.2f}%', fontsize=9)

    plt.xticks(np.arange(1, bins+1), [f'{i/bins*100:.0f}%' for i in range(1, bins+1)])
    plt.show()
```

```python
def train_model(model, X_train, y_train, patience, epochs, batch_size):
    early_stopping = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
    model.summary()
    history = model.fit(X_train, y_train, epochs=300, batch_size=32, validation_split=0.1, callbacks=[early_stopping])
    return history
```

```python
def NLL(y, distr): 
    return -distr.log_prob(y) 

def normal_sp(params): 
    return tfd.Normal(loc=params[:, 0:1], scale=1e-3 + tf.math.softplus(0.05 * params[:, 1:2])) # both parameters are learnable
```

### Deep Neural Network with Gaussian Output

```python
def create_model_mlp_gaussian(X_train, y_train, epochs=100, batch_size=32, validation_split=0.1):
    inputs = Input(shape=(X_train.shape[1],))
    hidden = Dense(80, activation="relu")(inputs)
    hidden = Dense(50, activation="relu")(hidden)
    hidden = Dense(20, activation="relu")(hidden)
    params = Dense(2)(hidden)
    dist = tfp.layers.DistributionLambda(normal_sp)(params) 

    model_mlp_gaussian = Model(inputs=inputs, outputs=dist)
    model_mlp_gaussian.compile(Adam(learning_rate=0.001), loss=NLL)

    return model_mlp_gaussian

model_mlp_gaussian = create_model_mlp_gaussian(X_train, y_train)
history = train_model(model_mlp_gaussian, X_train, y_train, 10, 100, 32)
```

### CNN + Deep Neural Network with Gaussian Output

```python
X_train_cnn = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test_cnn = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
def create_model_cnn_gaussian(X_train, y_train, epochs=100, batch_size=32, validation_split=0.1):
    inputs = Input(shape=(X_train_cnn.shape[1], 1))
    conv1d_layer = Conv1D(filters=64, kernel_size=5, activation='relu')(inputs)
    maxpooling_layer = MaxPooling1D(pool_size=2)(conv1d_layer)
    flatten_layer = Flatten()(maxpooling_layer)
    hidden1 = Dense(20, activation="relu")(flatten_layer)
    hidden2 = Dense(50, activation="relu")(hidden1)
    hidden3 = Dense(20, activation="relu")(hidden2)
    params = Dense(2)(hidden3)
    dist = tfp.layers.DistributionLambda(normal_sp)(params) 

    model_cnn_gaussian = Model(inputs=inputs, outputs=dist)
    model_cnn_gaussian.compile(Adam(learning_rate=0.01), loss=NLL)
    
    return model_cnn_gaussian

model_cnn_gaussian = create_model_cnn_gaussian(X_train_cnn, y_train)
history = train_model(model_cnn_gaussian, X_train, y_train, 20, 300, 32)
```

```python
def train_multivariate_model(model, X_train, y_train, epochs=100, batch_size=32, validation_split=0.1, patience=10):
    early_stopping = EarlyStopping(monitor='val_loss', patience=patience, restore_best_weights=True)

    model.build(X_train.shape)
    model.summary()

    history = model.fit(
        X_train, y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_split=validation_split,
        callbacks=[early_stopping]
    )

    return history
```

```python
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras.models import Model
import tensorflow_probability as tfp
import tensorflow as tf
from tensorflow.keras.optimizers import Adam

def multivariate_gaussian_with_covariance(d, input_size):
    inputs = Input(shape=input_size)

    h_means1 = Dense(80, activation='relu')(inputs)
    h_cov1 = Dense(80, activation='relu')(inputs)
    
    h_means2 = Dense(50, activation='relu')(h_means1)
    h_cov2 = Dense(50, activation='relu')(h_cov1)
    
    h_means3 = Dense(10, activation='relu')(h_means2)
    h_cov3 = Dense(10, activation='relu')(h_cov2)
    
    print(d)
    
    out_mean = Dense(d, activation='linear')(h_means3)  # Using linear activation for mean
    out_cov = Dense(tfp.layers.MultivariateNormalTriL.params_size(d) - d)(h_cov3)

    print(out_mean.shape)
    print(out_cov.shape)
    
    
    # Concatenate mean and lower triangular part of the covariance matrix
    concatenated_outputs = Concatenate()([out_mean, out_cov])
    
    print(concatenated_outputs.shape)
    
    # Define the distribution layer
    distribution_layer = tfp.layers.DistributionLambda(
        lambda t: tfd.MultivariateNormalTriL(
            loc=t[..., :d],
            scale_tril=tfp.math.fill_triangular(t[..., d:])
        )
    )
    
    model = Model(inputs=inputs, outputs=distribution_layer(concatenated_outputs),
                  name="multivariate_gaussian_with_covariance")
    
    model.compile(Adam(learning_rate=0.01),
                  loss=lambda y, model: -model.log_prob(y))
   
    return model
```

```python
'''
Output is a multivariate Gaussian with only diagonal covariance matrix. 
We use separate both hidden layers and CNN layers to predict separate
means and variances.
'''
def multivariate_gaussian_only_diagonal_separate_cnn(d, input_size):
    inputs = Input(shape=input_size)
    
    # Split the input into D sets
    input_splits = tf.split(inputs, num_or_size_splits=d, axis=1)

    separatesL1 = []
    separatesL2 = []
    outputs = []

    for i in range(d):
        # Apply separate convolutional layers to each set
        conv1d_layer = Conv1D(filters=64, kernel_size=5, activation='relu')(input_splits[i])
        maxpooling_layer = MaxPooling1D(pool_size=2)(conv1d_layer)
        flatten_layer = Flatten()(maxpooling_layer)
        
        # Use separate hidden layers
        separatesL1.append(Dense(30, activation="relu")(flatten_layer))

    for i in range(d):
        separatesL2.append(Dense(10, activation="relu")(separatesL1[i]))

    for i in range(d):
        outputs.append(Dense(2)(separatesL2[i]))

    concatenated_outputs = Concatenate()([output for output in outputs])
    distribution_layer = tfp.layers.DistributionLambda(
        lambda t: tfd.MultivariateNormalDiag(loc=t[..., 0::2], scale_diag=t[..., 1::2])
    )
    model_nobay_multivariate = Model(inputs=inputs, outputs=distribution_layer(concatenated_outputs),
                                    name="multivariate_gaussian_only_diagonal_separate_cnn")
    model_nobay_multivariate.compile(Adam(learning_rate=0.01),
                                     loss=lambda y, model: -model.log_prob(y)
                                    )
    
    return model_nobay_multivariate
```

```python
model_multivariate_gaussian_only_diagonal_separate_cnn = multivariate_gaussian_only_diagonal_separate_cnn(d, (X_train_cnn.shape[1], 1))
history = train_multivariate_model(model_multivariate_gaussian_only_diagonal_separate_cnn, X_train_cnn, y_train_2)
```
