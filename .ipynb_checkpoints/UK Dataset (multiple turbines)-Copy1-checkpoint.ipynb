#### Auxiliary Functions

```python
def plot_means_variances(y_true, y_means, y_stddevs):
    min_vals = np.min([np.min(y_true), np.min(y_means)])
    max_vals = np.max([np.max(y_true), np.max(y_means)])

    plt.figure(figsize=(16, 6))

    # Plot means
    plt.subplot(1, 2, 1)
    plt.scatter(y_true, y_means, alpha=0.5, s=5)
    plt.plot([min_vals, max_vals], [min_vals, max_vals], 'k--')  # Add diagonal line
    plt.title('Means')
    plt.xlabel('True Values')
    plt.ylabel('Predicted Values')

    # Plot standard deviations
    plt.subplot(1, 2, 2)
    plt.scatter(y_true, y_stddevs, alpha=0.5, s=5)
    plt.title('Standard Deviation')
    plt.xlabel('True Values')
    plt.ylabel('Predicted Standard Deviation')

    plt.tight_layout()
    plt.show()

    
def evaluate_and_print_metrics(y_train, y_test, y_train_pred, y_test_pred, y_train_stddevs, y_test_stddevs):

    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

    train_made = mean_absolute_error(y_train, y_train_pred)
    test_made = mean_absolute_error(y_test, y_test_pred)

    train_lower_bound = y_train_pred - 1.96 * y_train_stddevs
    train_upper_bound = y_train_pred + 1.96 * y_train_stddevs

    test_lower_bound = y_test_pred - 1.96 * y_test_stddevs
    test_upper_bound = y_test_pred + 1.96 * y_test_stddevs

    train_within_interval = np.sum(np.logical_and(y_train.ravel() >= train_lower_bound, y_train.ravel() <= train_upper_bound))
    test_within_interval = np.sum(np.logical_and(y_test.ravel() >= test_lower_bound, y_test.ravel() <= test_upper_bound))

    train_percentage_within_interval = (train_within_interval / len(y_train.ravel())) * 100
    test_percentage_within_interval = (test_within_interval / len(y_test.ravel())) * 100

    # Print the results
    print(f"Train RMSE: {train_rmse:.4f}")
    print(f"Test RMSE: {test_rmse:.4f}")
    print(f"Train MAE: {train_made:.4f}")
    print(f"Test MAE: {test_made:.4f}")
    print(f"Percentage of Train Data Points within 95% CI: {train_percentage_within_interval:.2f}%")
    print(f"Percentage of Test Data Points within 95% CI: {test_percentage_within_interval:.2f}%")
```

```python
def plot_binned_averages(y_pred, y_stddev, num_bins=50):
    # Create an array of bin edges
    bin_edges = np.linspace(min(y_pred), max(y_pred), num_bins + 1)

    # Calculate the means and standard deviations within each bin
    bin_means = []
    bin_stddevs = []

    for i in range(num_bins):
        mask = (y_pred >= bin_edges[i]) & (y_pred < bin_edges[i + 1])
        if np.any(mask):
            bin_means.append(np.mean(y_pred[mask]))
            bin_stddevs.append(np.mean(y_stddev[mask]))
        else:
            bin_means.append(0)  # Placeholder for empty bins
            bin_stddevs.append(0)

    # Create a scatter plot of binned averages
    plt.figure(figsize=(8, 6))
    plt.scatter(bin_means, bin_stddevs, c='b', marker='o', alpha=0.5)
    plt.title('Binned Averages of Means vs. Standard Deviations')
    plt.xlabel('Binned Mean')
    plt.ylabel('Average Standard Deviation')
    plt.grid(True)
    plt.show()
```

```python
def plot_scatter_for_feature(X, y, model, feature_index, feature_name):
    # Number of samples to plot
    num_samples_to_plot = 1000

    # Randomly sample 1000 data points
    sample_indices = np.random.choice(X.shape[0], num_samples_to_plot, replace=False)
    X_samples = X[sample_indices, [feature_index]]
    y_samples = np.array(y)[sample_indices]

    # Calculate the predicted means for the samples
    y_model = model(X)
    y_means = np.array(y_model.mean())[sample_indices]

    # Create the scatter plot
    plt.figure(figsize=(12, 8))
    plt.scatter(X_samples, y_samples, label='True Values', alpha=0.4)
    plt.scatter(X_samples, y_means, label='Predicted Means', color='red', alpha=0.4)
    plt.xlabel(feature_name)
    plt.ylabel('Output')
    plt.legend()
    plt.title(f'Scatter Plot of True Values and Predicted Means ({feature_name})')
    plt.show()
```

```python
def plot_confidence_interval_histogram(y_test_pred, y_test_std, y_test, bins=20):
    z_scores = np.array([norm.ppf(i/bins + (1-i/bins)/2) for i in range(1, bins+1)])

    percentages_within_interval = []

    for z_score in z_scores:
        lower_bounds = y_test_pred.ravel() - z_score * y_test_std
        upper_bounds = y_test_pred.ravel() + z_score * y_test_std

        is_within_interval = np.logical_and(y_test >= lower_bounds, y_test <= upper_bounds)
        num_within_interval = np.sum(is_within_interval)

        percentage_within_interval = (num_within_interval / len(y_test)) * 100
        percentages_within_interval.append(percentage_within_interval)

    plt.figure(figsize=(16, 6))
    bars = plt.bar(np.arange(1, bins+1), percentages_within_interval, color='skyblue', edgecolor='black')
    plt.xlabel('Confidence Intervals')
    plt.ylabel('Percentage within Interval')
    plt.title('Percentage of Data Points within Confidence Intervals')

    # Add numbers on top of each bar
    for bar, percentage in zip(bars, percentages_within_interval):
        plt.text(bar.get_x() + bar.get_width() / 2 - 0.15, bar.get_height() + 1, f'{percentage:.2f}%', fontsize=9)

    plt.xticks(np.arange(1, bins+1), [f'{i/bins*100:.0f}%' for i in range(1, bins+1)])
    plt.show()
```

```python
def flatten_dataframe(df, features, output_feature, date_col):
    df['turbine_id'] = df['turbine']
    df = df.pivot(index=date_col, columns='turbine_id', values=features + [output_feature])
    df.columns = ['_'.join(map(str, col)).strip() for col in df.columns.values]
    return df
```

```python
def show_corr_matrix(selected_columns):
    df_selected = df[selected_columns]
    correlation_matrix = df_selected.corr()

    plt.figure(figsize=(16, 13))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Matrix')
    plt.show()
```

```python
def univariate_dataframe(df_flatten, turbine_id_predicted, features, output_feature):
    turbine_id_predicted = '1'
    output_feature_id = f'{output_feature}_{turbine_id_predicted}' 
    other_power_columns = [col for col in df_flatten.columns if col.startswith(output_feature) and col != output_feature_id]
    features_id = ['_'.join([f, turbine_id_predicted]) for f in features]
    columns = [output_feature_id] + other_power_columns + features_id
    df_with_other_powers = df_flatten[columns].dropna()
    return df_with_other_powers, other_power_columns + features_id, output_feature_id
```

```python
def plot_epochs_loss_history(history):
    plt.plot(history.history['loss'][1:])
    plt.legend(['loss'])
    plt.ylabel('NLL')
    plt.xlabel('Epochs')
    plt.show()
```

```python
def plot_means_variances_multivariate(y_true, y_means, y_stddevs):
    d = y_true.shape[1]  # Number of dimensions
    
    fig, axs = plt.subplots(1, d, figsize=(20, 10))

    min_vals = np.min([np.min(y_true), np.min(y_means)])
    max_vals = np.max([np.max(y_true), np.max(y_means)])

    for i in range(d):
        axs[i].scatter(y_true[:, i], y_means[:, i], label=f'Column {i + 1}', alpha=0.5, s=10)
        axs[i].plot([min_vals, max_vals], [min_vals, max_vals], 'k--')  # Add diagonal line
        axs[i].set_title(f'Means - Dimension {i + 1}')
        axs[i].set_xlabel(f'True Values - Dimension {i + 1}')
        axs[i].set_ylabel(f'Predicted Values - Dimension {i + 1}')
        axs[i].legend()

    plt.show()

    fig, axs = plt.subplots(1, d, figsize=(20, 10))

    for i in range(d):
        axs[i].scatter(y_true[:, i], np.sqrt(y_stddevs[:, i]), label=f'Column {i + 1}', alpha=0.5, s=10)
        axs[i].set_title(f'Standard Deviation - Turbine {i + 1}')
        axs[i].set_xlabel(f'True Values - Turbine {i + 1}')
        axs[i].set_ylabel(f'Predicted Standard Deviation - Turbine {i + 1}')
        axs[i].legend()

    plt.show()
```

```python
def plot_binned_averages_multivariate(y_true, y_means, y_stddevs, num_bins=50):
    d = y_true.shape[1]  # Number of dimensions

    bin_edges = [np.linspace(min(y_true[:, i]), max(y_true[:, i]), num_bins + 1) for i in range(d)]

    bin_means = [[] for _ in range(d)]
    bin_stddevs = [[] for _ in range(d)]

    for i in range(d):
        for j in range(num_bins):
            mask = (y_true[:, i] >= bin_edges[i][j]) & (y_true[:, i] < bin_edges[i][j + 1])
            if np.any(mask):
                bin_means[i].append(np.mean(y_means[mask, i]))
                bin_stddevs[i].append(np.mean(y_stddevs[mask, i]))
            else:
                bin_means[i].append(0)
                bin_stddevs[i].append(0)

    fig, axs = plt.subplots(1, d, figsize=(20, 10))

    for i in range(d):
        axs[i].scatter(bin_means[i], bin_stddevs[i], c='b', marker='o', alpha=0.5)
        axs[i].set_title(f'Binned Averages - Turbine {i + 1}')
        axs[i].set_xlabel(f'Binned Mean - Turbine {i + 1}')
        axs[i].set_ylabel(f'Average Standard Deviation - Turbine {i + 1}')
        axs[i].grid(True)

    plt.show()
```

```python
def plot_confidence_interval_histogram_multivariate(y_test_pred, y_test_std, y_test, bins=20):
    d = y_test.shape[1]  # Number of dimensions
    z_scores = np.array([norm.ppf(i / bins + (1 - i / bins) / 2) for i in range(1, bins + 1)])

    percentages_within_interval = {f'Turbine {i + 1}': [] for i in range(d)}

    for z_score in z_scores:
        for i in range(d):
            lower_bounds = y_test_pred[:, i] - z_score * y_test_std[:, i]
            upper_bounds = y_test_pred[:, i] + z_score * y_test_std[:, i]

            is_within_interval = np.logical_and(y_test[:, i] >= lower_bounds, y_test[:, i] <= upper_bounds)
            num_within_interval = np.sum(is_within_interval)

            percentage_within_interval = (num_within_interval / len(y_test)) * 100
            percentages_within_interval[f'Turbine {i + 1}'].append(percentage_within_interval)

    plt.figure(figsize=(16, 18))

    for i in range(d):
        plt.subplot(d, 1, i + 1)
        bars = plt.bar(np.arange(1, bins + 1), percentages_within_interval[f'Turbine {i + 1}'], color='skyblue',
                       edgecolor='black')
        plt.xlabel('Confidence Intervals')
        plt.ylabel('Percentage within Interval')
        plt.title(f'Percentage of Data Points within Confidence Intervals - Turbine {i + 1}')

        # Add numbers on top of each bar
        for bar, percentage in zip(bars, percentages_within_interval[f'Turbine {i + 1}']):
            plt.text(bar.get_x() + bar.get_width() / 2 - 0.15, bar.get_height() + 1, f'{percentage:.2f}%', fontsize=9)

        plt.xticks(np.arange(1, bins + 1), [f'{j / bins * 100:.0f}%' for j in range(1, bins + 1)])

    plt.tight_layout()
    plt.show()
```

```python
def NLL(y, distr): 
    return -distr.log_prob(y) 

def normal_sp(params): 
    return tfd.Normal(loc=params[:, 0:1], scale=1e-3 + tf.math.softplus(0.05 * params[:, 1:2])) # both parameters are learnable
```

## Non-Bayesian
We create non-Bayesian deep models predicting the mean and the standard deviation of univariate and multivariate Gaussian distribution by minimizing the negative log-likelihood of the distribution final layer. We do so using a deterministic deep learning model. Since we have sufficient data covering the whole feature space using bayesian layers would be an overhead so the model has deterministic weights and probabilistic output (we only model aleatoric uncertainty and not epistemic).

### 1. Univariate Gaussian Output
The output distribution is a univariate Gaussian based off only a mean and standard deviation unit.

#### 1.1. Deep Dense Neural Network with (Univariate) Gaussian Output

```python
inputs = Input(shape=(X_train.shape[1],))
hidden = Dense(20, activation="relu")(inputs)
hidden = Dense(50, activation="relu")(hidden)
hidden = Dense(20, activation="relu")(hidden)
params = Dense(2)(hidden)
dist = tfp.layers.DistributionLambda(normal_sp)(params) 

model_nobay = Model(inputs=inputs, outputs=dist)
model_nobay.compile(Adam(), loss=NLL)

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

model_nobay.summary()

history = model_nobay.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1, callbacks=[early_stopping])
```

### 2. Multivariate Gaussian
In this case we are predicting the distribution of the power output of multiple turbines via a mean vector and a covariance matrix.

https://www.tensorflow.org/probability/api_docs/python/tfp/layers/MultivariateNormalTriL#example

https://rpubs.com/bacti/tensorflow-probabilistic

```python
def train_multivariate_model(model, X_train, y_train, epochs=100, batch_size=32, validation_split=0.1, patience=10):
    early_stopping = EarlyStopping(monitor='val_loss', patience=patience, restore_best_weights=True)

    model.build(X_train.shape)
    model.summary()

    history = model.fit(
        X_train, y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_split=validation_split,
        callbacks=[early_stopping]
    )

    return history
```

#### 2.1. CNN + Deep Dense Neural Network with (Multivariate) Gaussian Output using unconstrained Lower-Triangular Covariance Matrix
Output is a multivariate Gaussian with full lower triangular covariance matrix. We use same hidden layers to predict separate means, variances and covariances.

```python
'''
Output is a multivariate Gaussian with full lower triangular covariance matrix. 
We use same hidden layers to predict separate means, variances and covariances.
'''
def multivariate_gaussian_with_covariance(d):
    model_nobay_multivariate = Sequential([
        Conv1D(filters=64, kernel_size=5, activation='relu'),
        MaxPooling1D(pool_size=2),
        Flatten(),
        Dense(50, activation="relu"),
        Dense(tfp.layers.MultivariateNormalTriL.params_size(d)),
        tfp.layers.MultivariateNormalTriL(
            d
        )

    ],
    name = "multivariate_gaussian_with_covariance"
    )
    model_nobay_multivariate.compile(Adam(learning_rate=0.01),
                                     loss=lambda y, model: -model.log_prob(y)
                                    )
   
    return model_nobay_multivariate
```

```python
model_multivariate_gaussian_with_covariance = multivariate_gaussian_with_covariance(d)
history = train_multivariate_model(model_multivariate_gaussian_with_covariance, X_train_cnn, y_train)
```

#### 2.2. CNN + Deep Dense Neural Network with (Multivariate) Gaussian Output using only diagonal Covariance Matrix

Output is a multivariate Gaussian with only diagonal covariance matrix. We use same hidden layers to predict separate means and variances.

#### 2.3. CNN + Deep Dense Neural Network with (Multivariate) Gaussian Output using only diagonal Covariance Matrix - but using partially separate hidden dense layers 
Output is a multivariate Gaussian with only diagonal covariance matrix.  We use separate hidden dense layers to predict separate means and variances. In this case, we include an initial shared convolutional layer and a maxpooling layer.

```python
'''
Output is a multivariate Gaussian with only diagonal covariance matrix. 
We use separate hidden layers to predict separate means and variances.
In this case, we include an initial shared convolutional layer and a
maxpooling layer.
'''
def multivariate_gaussian_only_diagonal_separate_layers(d):
    inputs = Input(shape=(X_train_cnn.shape[1], 1))

    conv1d_layer = Conv1D(filters=64, kernel_size=5, activation='relu')(inputs)

    maxpooling_layer = MaxPooling1D(pool_size=2)(conv1d_layer)

    flatten_layer = Flatten()(maxpooling_layer)

    separatesL1 = []
    separatesL2 = []
    outputs = []

    for i in range(d):
        separatesL1.append(Dense(30, activation="relu")(flatten_layer))

    for i in range(d):
        separatesL2.append(Dense(10, activation="relu")(separatesL1[i]))

    for i in range(d):
#         x1 = (Dense(1, activation='linear')(separatesL2[i]))
#         x2 = (Dense(1, activation='softplus')(separatesL2[i]))
#         outputs.append(x1)
#         outputs.append(x2)
        outputs.append(Dense(2)(separatesL2[i]))

    concatenated_outputs = Concatenate()([output for output in outputs])
    distribution_layer = tfp.layers.DistributionLambda(
        lambda t: tfd.MultivariateNormalDiag(loc=t[..., 0::2], scale_diag=t[..., 1::2])
    )
    model_nobay_multivariate = Model(inputs=inputs, outputs=distribution_layer(concatenated_outputs),
                                    name = "multivariate_gaussian_only_diagonal_separate_layers")
    model_nobay_multivariate.compile(Adam(learning_rate=0.001),
                                     loss=lambda y, model: -model.log_prob(y)
                                    )
    
    return model_nobay_multivariate
```

#### 2.4. CNN + Deep Dense Neural Network with (Multivariate) Gaussian Output using only diagonal Covariance Matrix - but using separate CNN and hidden dense layers
Output is a multivariate Gaussian with only diagonal covariance matrix.  We use separate both CNN and hidden dense layers to predict separate means and variances.

```python
'''
Output is a multivariate Gaussian with only diagonal covariance matrix. 
We use separate both hidden layers and CNN layers to predict separate
means and variances.
'''
def multivariate_gaussian_only_diagonal_separate_cnn(d):
    inputs = Input(shape=(X_train_cnn.shape[1], 1))
    
    # Split the input into D sets
    input_splits = tf.split(inputs, num_or_size_splits=d, axis=1)

    separatesL1 = []
    separatesL2 = []
    outputs = []

    for i in range(d):
        # Apply separate convolutional layers to each set
        conv1d_layer = Conv1D(filters=64, kernel_size=5, activation='relu')(input_splits[i])
        maxpooling_layer = MaxPooling1D(pool_size=2)(conv1d_layer)
        flatten_layer = Flatten()(maxpooling_layer)
        
        # Use separate hidden layers
        separatesL1.append(Dense(30, activation="relu")(flatten_layer))

    for i in range(d):
        separatesL2.append(Dense(10, activation="relu")(separatesL1[i]))

    for i in range(d):
        outputs.append(Dense(2)(separatesL2[i]))

    concatenated_outputs = Concatenate()([output for output in outputs])
    distribution_layer = tfp.layers.DistributionLambda(
        lambda t: tfd.MultivariateNormalDiag(loc=t[..., 0::2], scale_diag=t[..., 1::2])
    )
    model_nobay_multivariate = Model(inputs=inputs, outputs=distribution_layer(concatenated_outputs),
                                    name="multivariate_gaussian_only_diagonal_separate_cnn")
    model_nobay_multivariate.compile(Adam(learning_rate=0.01),
                                     loss=lambda y, model: -model.log_prob(y)
                                    )
    
    return model_nobay_multivariate
```

```python
model_multivariate_gaussian_only_diagonal_separate_cnn = multivariate_gaussian_only_diagonal_separate_cnn(d)
history = train_multivariate_model(model_multivariate_gaussian_only_diagonal_separate_cnn, X_train_cnn, y_train)
```
